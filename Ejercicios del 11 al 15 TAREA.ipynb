```python
num1= int(input("Ingrese su primer número"))
num2= int(input("Ingrese su segundo número"))
operación= input("¿Qué operación desea realizar?( +, -, *, /)")
if operación =="*": print(num1*num2)
elif operación =="+": print(num1+num2)
elif operación =="/": print(num1/num2)
elif operación =="-": print(num1-num2)
else: print("lo que ingreso no corresponde con las opciones dadas porfavor vuleva a intentar")
```

```python
num1 = int(input("Ingrese su primer número: "))
num2 = int(input("Ingrese su segundo número: "))
operacion = input("¿Qué operación desea realizar? (+, -, *, /): ")

resultado = (
    num1 * num2 if operacion == "*" else
    num1 + num2 if operacion == "+" else
    num1 / num2 if operacion == "/" and num2 != 0 else
    "Error: División entre 0" if operacion == "/" and num2 == 0 else
    num1 - num2 if operacion == "-" else
    "Opción no válida"
)

print(resultado)
```
