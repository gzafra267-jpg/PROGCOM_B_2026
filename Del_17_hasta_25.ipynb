```python
n = int(input("Ingrese un número natural: "))
suma = 0
for i in range(1, n + 1):
    suma += i

print("La suma de los números naturales hasta", n, "es:", suma)
```

```python
n = int(input("Ingrese la cantidad de números que desea ver: "))

a = 0
b = 1
contador = 0

print("Serie Fibonacci:")

while contador < n:
    print(a)
    siguiente = a + b
    a = b
    b = siguiente
    contador += 1
```

```python
import random

colores = []

print("Escriba 10 colores:")

for i in range(10):
    color = input("Color: ")
    colores.append(color)

secreto = random.choice(colores)

print("Tiene 10 intentos para adivinar el color.")

for i in range(10):
    intento = input("Adivine el color: ")

    if intento == secreto:
        print("Adivinó el color")
        break
    else:
        print("No es ese color")

if intento != secreto:
    print("Se acabaron los intentos")
    print("El color era:", secreto)
```

```python
import random

colores = []

print("Escriba 10 colores:")

i = 0
while i < 10:
    color = input("Color: ")
    colores.append(color)
    i = i + 1

# Elegir una posición al azar entre 0 y 9
posicion = random.randint(0, 9)
secreto = colores[posicion]

print("Tiene 10 intentos para adivinar el color.")

intentos = 0

while intentos < 10:
    intento = input("Adivine el color: ")
    intentos = intentos + 1

    if intento == secreto:
        print("Adivinó el color")
        break
    else:
        print("No es ese color")

if intento != secreto:
    print("Se acabaron los intentos")
    print("El color era:", secreto)
```

```python
numero = int(input("Ingrese un número: "))

suma = 0

while numero > 0:
    digito = numero % 10
    suma = suma + digito
    numero = numero // 10

print("La suma de los dígitos es:", suma)
```

```python
numero = input("Ingrese un número: ")
suma = 0

for digito in numero:
    suma = suma + int(digito)

print("La suma de los dígitos es:", suma)
```

```python
numero = int(input("Ingrese un número: "))

i = 1
contador = 0

while i <= numero:
    if numero % i == 0:
        contador = contador + 1
    i = i + 1

if contador == 2:
    print("El número es primo")
else:
    print("El número no es primo")
```

```python
filas = int(input("Ingrese el número de filas: "))

i = 1

while i <= filas:
    j = 1
    while j <= i:
        print("*", end="")
        j = j + 1
    print()   # salto de línea
    i = i + 1
```
