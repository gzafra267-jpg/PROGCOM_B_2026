#Programació orientada a objetos POO
- Paradigma de programación enfocado en crear marcos(clase) que a su vez genera instancias(objetos).
- Está funadamentada bajo 4 principios fundamentales o pilares: **Herencia,
Abstracción, Polimorfismo y Encapsulamiento .**

-Se enfoca en crear plantillas (class) las cuales desarrollan instancias(objetios). Esos objetos hacen (métodos), y tienen caracaterísticas propias(atributos).

- Ventajas al implementar POO
1. Escalabilidad: Un mismo código en diferentes apps.
2. Seguridad: No cualquier persona tiene acceso al  código.
3. Modularidad:El codigo se epresa a partir de "pedazos" (métodos)
4. Reusabilidad: Se puede reutilizar partes del código en lugar de generar un código nuevo.
5. Mantenibilidad: El codigo puede ser fácilmente corregido y actualizado.


````python
class Edificio:
  # Contructor: para crear atributos,__init__/(parámetros)
  def __init__(self,nombre,habitación,dirección):
    self.nombre=nombre
    self.habitación=habitación
    self.dirección=dirección
    #MÉTODOS SON LASMISMAS FUNCIONES SOLO QUE ESTAN DENTRO DE CLASS
  def mostrar_info(self):
   print(f"Nombre:{self.nombre}")
   print(f"Dirección: {self.dirección}")
   print(f"No. de habitaciones {self.habitación}")

class floristeria(Edificio):
  def __init__(self,nombre,habitación,dirección,tipo_flores):
   super().__init__(nombre,habitación,dirección)
   self.tipo_flores=tipo_flores
  def vender_flor(self,flor):
    if flor in self.tipo_flores:
      print(f"Se vendió una flor de tipo{flor}")
    else:
      print(f"No tengo de esa flor")

````

````python
mi_floristeria= floristeria("Floristería El chamo",4,"carrera 33, 28-15 Bucaramanga",["rosas","tilipán","lirios"])
mi_floristeria.mostrar_info()
````
