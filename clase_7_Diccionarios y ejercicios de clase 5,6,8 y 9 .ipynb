crea un diccionario con nombres de estudiantes y sus notas Muestra el pormedio de todos los estudientes y de un estudiante especifico, empleando dos funciones


estudiantes = { "Ana": [4.5, 3.8, 4.2], "Luis": [3.0, 3.5, 4.0], "María": [5.0, 4.8, 4.9],"Carlos": [2.8, 3.2, 3.5]}


def promedio_general(diccionario):
    suma_total = 0
    cantidad_notas = 0
    
    for notas in diccionario.values():
        suma_total += sum(notas)
        cantidad_notas += len(notas)
    
    return suma_total / cantidad_notas


def promedio_estudiante(diccionario, nombre):
    if nombre in diccionario:
        notas = diccionario[nombre]
        return sum(notas) / len(notas)
    else:
        return "Estudiante no encontrado"

print("Promedio general:", promedio_general(estudiantes))
print("Promedio de María:", promedio_estudiante(estudiantes, "María"))

dada una lista de elementos recolectandos en una expedición, cuántas vece aparece cada elemento
lista sugerida: ["gema","oro","gema","madera","oro","oro","gema"]

```python

elementos = ["gema","oro","gema","madera","oro","oro","gema"]

conteo = {}

for elemento in elementos:
    if elemento in conteo:
        conteo[elemento] += 1
    else:
        conteo[elemento] = 1

for clave, valor in conteo.items():
    print(clave, ":", valor)
```

desarrolla el ejercicio de la floristería (el desarrollador con lista) pero empleado diccionario

```python

ventas = ["rosa", "tulipan", "rosa", "girasol", "rosa", "tulipan"]

floristeria = {}

for flor in ventas:
    if flor in floristeria:
        floristeria[flor] += 1
    else:
        floristeria[flor] = 1

print("Ventas por tipo de flor:")
for flor, cantidad in floristeria.items():
    print(flor, ":", cantidad)
```
